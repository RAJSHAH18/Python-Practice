```python
import pandas as pd

a = pd.read_csv("weather.csv", encoding="latin-1")
a.head(10)
```

```python
# Find all the unique “wind speed” values in the data.

tmp = a["Wind Speed_km/h"].unique()
print(tmp)
```

```python
# Find a number of times when the “wind speed” was exactly 4 km/h.

# tp = a["Wind Speed_km/h"]
# count = 0;

# if tp==4:
#     count = count + 1
    
# print(count)
s=a[a['Wind Speed_km/h'] == 4]
print(s)
```

```python
# Find out all the null values in the data.

demo = a.isnull()
print(demo)
```

```python
# Find all the instances of when “wind speed  is above 24”  and “visibility is 25”.


inst = a[(a['Wind Speed_km/h']>24) & (a['Visibility_km'] == 25)]
print(inst)
```

```python
# find the mean value of each column against each “weather condition”

mn = a.groupby("Weather").mean()
print(mn)
```

```python
# Find the min and max value of each column against each  “weather condition”

mx = a.groupby('Weather').max()
mn = a.groupby('Weather').min()
print(mx)
print(mn)
```

```python
# Find all the records where the weather condition is fog.

fg = a[a['Weather'] == 'Fog']
print(fg)
```

```python
# Find all the instances when ‘weather is clear’ or visibility is 40’.

st = a[(a['Weather'] == 'Clear') | (a['Visibility_km'] == 40)]
print(st)
```

```python
# Find the weather  is clear and relative humidity is greater than 50 or visibility is above 40.

print(a[(a['Weather']=='Clear')])
```

```python
# Find the top 10 hottest temp values and their counts.

hot_city = a.nlargest(10, 'Temp_C').count()
print(hot_city)
```

```python
# Find the number of times when the weather is exactly clear.

weth = a.groupby("Weather").get_group("Clear")
print(weth)
```
