# BAR GRAPH

```python
import matplotlib.pyplot as plt
x = ["Python","C","C++","JAVA"]
c = [85,70,60,82]
plt.xlabel("Language",fontsize=15)
# For providing font size used like this plt.xlabel("language",fontsize=15)
plt.ylabel("NUMBER")
plt.title("WSCUBE")
# To change width of bars in graph provide width parameter  in plt.bar(width = 0.4).
# To Provide Different colours to the graph Used color parameter also u provide same colour to each grap or different also
# c = ["red","Orange","Blue","Green"]
c = ["b","y","m","g"]
plt.bar(x,c,width = 0.4,color=['blue','red','blue','green'])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = ["Python","C","C++","JAVA"]
y = [85,70,98,82]
z = [40,30,50,20]
p = np.arange(len(x))
plt.xlabel("LANGAUAGE",fontsize = 15)
plt.ylabel("NUMBER",fontsize = 15)
plt.title("WSCUBE")
# align parameter is used in plt.bar(align ="edge/center") to set the alignment of header of x-axis and by default is center only
# to change colour boundary of bar used edgecolour like edgecolour = "r" and to provide thickness to it use linewidth=5 
# to change the style of change used linestyle= ":"[for dotted linestyle] by default is solid
# to want like coolur we used alpha parameter and it values lies between 0-1
# Used llabel paramter to bar graph used label parameter like label = "WSCUBE" and also for this u have to use legend like plt.legend()
# to provide common wodth used width=0.2

width = 0.2
p1 = [j + width for j in p]
plt.barh(p,y,width,color = ["red","blue","green","orange"],align = "center",edgecolor = "black",linewidth=3,alpha=0.5,label="Popularity")
plt.barh(p1,z,width,align = "center",color="y",label="Popularity1")

plt.legend()
plt.show()
```

# PIE CHART

```python
import matplotlib.pyplot as plt
x = ["Python","C","C++","JAVA"]
c = [85,70,60,82]
plt.xlabel("Language",fontsize=15)
# For providing font size used like this plt.xlabel("language",fontsize=15)
plt.ylabel("NUMBER")
plt.title("WSCUBE")
plt.pie(c,labels=x,autopct="%2.1f%%")
plt.legend()
plt.show()
```
