# BAR GRAPH

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [11,12,13,14,15,16]
plt.xlabel("Number",fontsize=15)
plt.ylabel("OBJECT",fontsize=15,color="yellow")
plt.title("BAR GRAPH",fontsize=15,color="red")
plt.bar(x,y,color=["red","blue"])
plt.show()
```

# PIE CHART

```python
x = ["python","C","C++"]
y = [90,80,50]
plt.pie(y,labels=x,autopct="%0.2f%%",shadow=True)
plt.title("Popularity of languages")
plt.show()
```

# Line Graph

```python
x = [1,2,3,4,5,6]
y = [11,12,13,14,15,16]
plt.plot(x,y,color="red",marker="o")
plt.show()
```

# WORD CLOUD

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
f = open("myfile.txt")
s = f.read()
# type(s)
cloud = WordCloud(background_color="Black").generate(s)
plt.imshow(cloud)
plt.axis("off")
plt.show()
```
