# 1.	Take input of given text file and read contents of it line by line. Find Top 5 words with highest occurance ans plot vertical bar graph for same. 


```python
# Uisng Dicitonary
import matplotlib.pyplot as plt
file = open("myfile.txt","r")
file.seek(0)
lines = file.readlines()
words = []
for line in lines:
    line = line.split()
    words.extend(line)

d1 = {}
for i in words:
    if i not in d1:
        d1[i.upper()] = words.count(i)
top_5 = dict(sorted(d1.items(),key=lambda x : x[1],reverse = True)[:5])

# word_list = [word[0] for word in top_words]
# frequency_list = [word[1] for word in top_words]

# Bar Graph
x = top_5.keys()
y = top_5.values()
plt.xlabel("WORDS",fontsize = 15)
plt.ylabel("OCCURENCE OF WORDS",fontsize = 15)
plt.title("Top 5 WORDS",fontsize= 15)
plt.bar(x,y,width=0.5)
plt.show()
```

```python
# PIE CHART
plt.pie(y,labels=x,textprops = {"fontsize":10},autopct = "%0.2f%%",shadow=True)
plt.title("TOP 5 WORDS")
plt.legend(title = "WORDS")
plt.show()
```

```python
# Line Chart
plt.plot(x,y,marker="o", linestyle="dashed",color="r",linewidth="2")
plt.xlabel("WORDS")
plt.ylabel("OCCURENCE OF WORDS")
plt.title("TOP 5 WORDS")

plt.show()
```

# 3.	Create python list for set of programming keywords. Count occurance of all programmig keywords in given codeblock and plot graph for same.

```python
import matplotlib.pyplot as plt
keywords = ['if', 'else', 'elif', 'while', 'for', 'for', 'def', 'class', 'return', 'import', 'from', 'try', 'except', 'finally', 'raise', 'assert', 'and', 'or', 'not', 'in', 'is', 'or', 'and']
words = {}
for i in keywords:
    if i not in words:
        words[i]=keywords.count(i)
words = dict(sorted(words.items(),key=lambda x : x[1],reverse = True))
keyword = words.keys()
count = words.values()



```

```python
# Bar Graph 
plt.bar(keyword,count,)
plt.xlabel("KEYWORDS",fontsize=20)
plt.ylabel("OCCURRENCE OF KEYWORDS",fontsize=20)
plt.title("FREQUENCY OF KEYWORDS",fontsize=20)
plt.xticks(rotation=90)
plt.show()
```
